```python
import os
from glob import glob
from parse_data import make_outerjoin_df, get_features_from_df
from dotenv import load_dotenv
load_dotenv()
data_fn = os.environ.get("DATA_PATH")
bps_fns = glob(os.path.join(os.environ.get("BLUEPRINT_DIR"), "*"))

outerjoin_df = make_outerjoin_df(data_fn)
features_per_bp = get_features_from_df(outerjoin_df, mode="all")
```

```python
def check_dimdetail_template_match(to_check_input:dict, template:dict):
    to_check_input = to_check_input.copy()

    eq_part_type = to_check_input["part_type"] == template["part_type"]
    eq_classification = to_check_input["classification"] == template["classification"]
    
    to_check_input.pop("part_type")
    to_check_input.pop("classification")
    alphabet_list_in = set(to_check_input.keys())
    alphabet_list_tp = set(cur["alphabet"] for cur in template["dimension_details"])
    ok_alphabets = all(elem in alphabet_list_tp for elem in alphabet_list_in)

    if all([eq_part_type, eq_classification, ok_alphabets]):
        return True
    else:
        return False

def make_dimdetail_based_template(dimdetails:dict, template:dict, default_value_mode:Literal["safe", "greedy"]="safe"):
    
    result = {}
    for key, value in template.items():
        if key == "dimension_details":
            dimension_details = []
            for cur in template[key]:
                if cur["alphabet"] not in dimdetails.keys():
                    if default_value_mode == "safe":
                        continue
                dimension_details.append({
                    'name': cur["name"],
                    'value': cur["value"],
                    'alphabet': cur["alphabet"],
                    'desc': cur["desc"]
                })
            result[key] = dimension_details
        else:
            result[key] = value
    
    return result
```
